```python
import pandas as pd
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
# Read dataset from csv file.
x = pd.read_csv('dataset.csv')
```

```python
x.head()
```

```python
# Fine-tuning the sector.
z = x['Sector'].drop_duplicates()
for i, j in enumerate(z):
    x['Sector'] = x['Sector'].replace(j, i)
```

```python
x.head()
```

```python
# Fine-tuning the location of project.
z = x['place of project'].drop_duplicates()
for i, j in enumerate(z):
    x['place of project'] = x['place of project'].replace(j,i)
```

```python
x.head()
```

```python
feature_df = x[['Sector', 'Budget', 'Skilled workers', 'Unskilled workers']]
target = x['Time required']
```

```python
reg = RandomForestRegressor(n_estimators=100)
reg.fit(feature_df, target)
```

```python
# Saving the trained Regressor

with open('Duration_Predictor.pickle','wb') as f :
    pickle.dump(reg, f)
```

```python
# Loading the saved Regressor

pickle_in = open('Duration_Predictor.pickle','rb')
reg = pickle.load(pickle_in)
```

```python
test_sector = input("Enter 0 for rural areas and 1 for urban areas : ")
test_budget = input("Enter budget amount(in Crores) : ")
test_skilled_workers_count = input("Enter number of skilled workers that are employed : ")
test_unskilled_workers_count = input("Enter number of unskilled workers that are employed : ")
```

```python
Feature = [test_sector, test_budget, test_skilled_workers_count, test_unskilled_workers_count]
print("Time required(in years)", reg.predict([Feature])[0])
```
